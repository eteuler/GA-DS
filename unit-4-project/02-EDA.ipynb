### I can already see some problem columns based on the min and max values as well as many missing values. Unfortnately all these data points have not always been collected in the history of Boxing, which has lead to many missing values.
### Age max is 2016 (can't be right) and the min is -74. Weight minimum and max seem unrealistic, as well as reach and height. Limits will need to be enforced on this data, to normalize it: 
##### Age limit: 18-59, an individual cannot legally become a professional boxer until he/she is 18. The oldest title winnner ever was 59 yrs old, so the max limit here for age will be 60.
##### Height Limit: 255 cm is about 8 feet so the max here cannot be realistic. Height max limit will be 214cm which is about 7 ft, and the minimum limit will be 137 cm which is about 4 and a half feet. 
##### Reach limit: A max of 456cm is around 14 ft, so that max cannot be realistic, also a minimum reach of 1 cm cannot be realistic either. A limit needs to be set for Boxer Reach. The highest boxer reach ever recorded was 84 inches, so the upper bound limit will be 215cm (a tad more than 84 inches). In boxing, anything under 60inches is considered "short reach" so we'll set the lower bound at 130 cms or about 51 inches. 
##### Lastly, Weight can be a bit harder to limit, because in the heaviest division (heavyweight) there isn't technically a weight maximum, so there will be no upper bound for weight. For the lower bound, the lowest weight class in boxing is the flyweight, where boxers may not weigh more than 106 pounds. So, here we will create the lower bound as 90lbs 

```python
# creating the bounds on the data, in order to clean/normalize the data

boxing = boxing[(boxing.height_A <= 214) & (boxing.height_A >= 137) &
      (boxing.height_B <= 214) & (boxing.height_B >= 137) &
      (boxing.weight_B >= 90) & (boxing.weight_A >= 90) &
      (boxing.age_A <= 60) & (boxing.age_A >= 18) &
      (boxing.age_B <= 60) & (boxing.age_B >= 18) &
      (boxing.reach_A <= 215) & (boxing.reach_A >= 130) &
      (boxing.reach_B <= 215) & (boxing.reach_B >= 130)] 
```

### Looks like there are some decision options that were not listed in the data dictionary: NWS, TD, PTS
##### NWS is a"Newspaper Decision" and it's is very out of date at this point. It was basically a decision made by the newspapers (often local newspapers that were biased toward a certain fighter) for a fight that ended as "No decision". Now points are used for fights that were "No Decision", and that is what PTS stands for in this column. 
##### TD is technical decision, and it only occurs when a fight must be ended because of a boxers use of a headbutt, this is determined by the referee. I would put this in the same boat as RTD and DQ as a fluky decisons that this data would struggle to predict properly. 

### Therfore, the next step will be to remove any RTD, DQ, TD, and NWS decisions from the data

```python
boxing = boxing[(boxing.decision != 'RTD') & (boxing.decision != 'DQ') & (boxing.decision != 'TD') & (boxing.decision != 'NWS')]
```

```python
boxing.info()
```

```python
# dropping of stance and judges points columns 
boxing = boxing.drop(['stance_A', 'stance_B', 'judge1_A', 'judge1_B', 'judge2_A', 'judge2_B', 'judge3_A', 'judge3_B'], axis=1)
```

#### The upcoming code adds "Diff" variables, which is (boxer A - boxer B), so negative number would mean boxer B had a larger (age, weight, height, reach) or was older than boxer A. 
#### Also created total fights variable so that KO percentage, win percentage, loss percentage, and draw percentage columns can be created
## NOTE: Boxer A is the Boxer with the higher ranking in this data, therfore the favored boxer 

```python
boxing['ageDiff'] = boxing.age_A - boxing.age_B
boxing['weightDiff'] = boxing.weight_A - boxing.weight_B
boxing['heightDiff'] = boxing.height_A - boxing.height_B
boxing['reachDiff'] = boxing.reach_A - boxing.reach_B 
boxing['total_bouts_A'] = boxing.won_A + boxing.lost_A
boxing['total_bouts_B'] = boxing.won_B + boxing.lost_B
boxing['WinPercent_A'] = (boxing.won_A/boxing.total_bouts_A) * 100
boxing['WinPercent_B'] = (boxing.won_B/boxing.total_bouts_B) * 100
boxing['KoPercent_A'] = (boxing.kos_A/boxing.total_bouts_A) * 100
boxing['KoPercent_B'] = (boxing.kos_B/boxing.total_bouts_B) * 100
```

```python
## dropping columns that will not be used, including the variables we no longer need because of diff columns 
## age, weight, height, reach

boxing = boxing.drop(['age_A', 'age_B', 'weight_A', 'weight_B', 'height_A', 'height_B', 'reach_A', 'reach_B'], axis=1)
```
