## Intermediate level

#### Count how many movies have each of the content ratings.

```python
# Answer: use of value_counts method to count all appearances of each rating
movies['content_rating'].value_counts()
```

#### Use a visualization to display that same data, including a title and x and y labels.

```python
# Answer:
# plotting the value counts dats, as a barplot using .plot(kind='bar')
movies['content_rating'].value_counts().plot(kind='bar')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Movies: Rating Distribution');
```

#### Convert the following content ratings to "UNRATED": NOT RATED, APPROVED, PASSED, GP.

```python
# Answer: using the .replace, replacing the instructed ratings to "UNRATED"
movies.content_rating.replace('NOT RATED', 'UNRATED', inplace=True)
movies.content_rating.replace('APPROVED', 'UNRATED', inplace=True)
movies.content_rating.replace('PASSED', 'UNRATED', inplace=True)
movies.content_rating.replace('GP', 'UNRATED', inplace=True)
```

```python
#displaying same boxplot as abaove but with replaced ratings
movies['content_rating'].value_counts().plot(kind='bar')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Movies: Rating Distribution');
```

#### Convert the following content ratings to "NC-17": X, TV-MA.

```python
# Answer: using .replace again to change X and TV-MA to NC-17 in the content_rating column
movies.content_rating.replace('X', 'NC-17', inplace=True)
movies.content_rating.replace('TV-MA', 'NC-17', inplace=True)
```

```python
#displaying the content rating bar graph again with the replaced column values
movies['content_rating'].value_counts().plot(kind='bar')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Movies: Rating Distribution');
```

#### If there are missing values: examine them, then fill them in with "reasonable" values.

```python
# Answer: code to display full row of data where content_rating = NaN
movies[movies.content_rating.isnull()]
```

```python
#use of the .loc method to locate specific movie ttiles, and then changing the value of content_rating for these films
# these were the films wit NAN as their rating so there should be no NAN values as oputput of this cell
movies.loc[movies.title == 'Butch Cassidy and the Sundance Kid', 'content_rating'] = "PG"
movies.loc[movies.title == 'Where Eagles Dare', 'content_rating'] = "PG"
movies.loc[movies.title == 'True Grit', 'content_rating'] = "G"

print(movies.isnull().sum())
```

#### Calculate the average star rating for movies 2 hours or longer, and compare that with the average star rating for movies shorter than 2 hours.

```python
# Answer: creation of two different datasets 
#one with movies of a duration two hours or longer and one with movies of a duration less than 2 hours (120 minutes)

movies_twoplus = movies[movies['duration'] >= 120]
movies_undertwo = movies[movies['duration'] < 120]
```

```python
print(movies_twoplus.star_rating.mean())
print(movies_undertwo.star_rating.mean())
#movies that are 2+ hours have a slightly larger star rating on average 
```

#### Calculate the average duration for each genre.

```python
# Answer: grouped by content ratings and then finding the mean duration based on unique genre
# PG-13 movies, on average, have the longest duration
movies.groupby('content_rating')['duration'].mean().sort_values(ascending=False)
```
